## Second step - Understand local rule

Each opperation like +, *, ** have your own local rule as process numbers and generate the backpropagation

First of all is necessary undestand that formula below because that'll be fundamental on local rule

$$
f'(x) = \lim_{h \to 0}   \frac{f(x+h) - f(x)}{h}
$$




### Addition - Local rule

#### Identify a function

$
w = 3\\
x = 7\\
f = w + x\\
f = 3 + 7\\
f = 10
$

#### Local derivates

$\frac{\partial z}{\partial w} = 1
\qquad
\frac{\partial z}{\partial x} = 1
$

why ∂z/∂x  is equal to 1?

and what'll happen if w was increase ↑?

#### increase w ↑

#### Actual Computation
$
w = 3.16\\
x = 7\\

f = 3.16 + 7\\

$

#### Now apply function:
$$
f'(x) = \lim_{h \to 0}   \frac{f(x+h) - f(x)}{h}
$$

What is h?

$
x = 3\\
h = 0.16\\
x + h = 3.16
$
